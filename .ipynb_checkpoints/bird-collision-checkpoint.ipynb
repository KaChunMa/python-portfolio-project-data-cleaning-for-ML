#Bird Collision Analysis

```python
#importing modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
#importing and previewing the chicago data
chicago = pd.read_csv('Chicago_collision_data.csv')
print(chicago.head())
chicago.info()
```

```python
#correcting data types
chicago['Genus'] = chicago['Genus'].astype('string')
chicago['Species'] = chicago['Species'].astype('string')
chicago['Date'] = chicago['Date'].astype('string')
chicago['Locality'] = chicago['Locality'].astype('string')
chicago.dtypes
```

```python
#creating year and month columns
chicago['Year'] = chicago['Date'].apply(lambda x: int(x[0:4]))
chicago['Month'] = chicago['Date'].apply(lambda x: int(x[5:7]))
print(chicago.head())
chicago.dtypes
```

```python
#some overall summery statistics
chicago.describe(include='all')
```

```python
#surface level exploration the data

#genus
print(12063/69784)
print(chicago['Species'][chicago['Genus'] == 'Melospiza'].unique())
print('\n')

#species
print(10133/69784)
print(chicago['Genus'][chicago['Species'] == 'albicollis'].unique())
print('\n')

#locality
print(36315/69784)
print(1 - 36315/69784)
print(36315 - (69784 - 36315))
print((36315 - (69784 - 36315)) / 69784)
print('\n')

#year and month
print(chicago['Month'].unique())
print(2016 - 1978)
```

Most common genus is Melospiza at 12063 / 69784 = 17.29%, Melospiza species include georiana, lincolnii and melodia.
Most common species is albicollis at 14.52%, part of the Zonotrichia genus.
Most common speices is not part of the the most common genus.

Ratio between localities MP and CHI is about the same at 52.04% and 47.96%, or counts 36315 and 33469 respectively, a difference of 2846 which is 4.08% of the total counts.

Min month is 3 and max month is 11, so the winter months (1, 2 and 12) and first two months of summer (6 and 7) have no bird collision in chicago over 38 years from 1978 to 2016.

```python
#counts by month and year
chicago_counts_monthBYyear = chicago.groupby(['Month','Year'])['Year'].count()
chicago_counts_monthBYyear = pd.DataFrame(chicago_counts_monthBYyear)
chicago_counts_monthBYyear.columns = ['Counts']
chicago_counts_monthBYyear
```

```python
#counts info by month across years
chicago_counts_month = chicago_counts_monthBYyear.groupby('Month')['Counts'].mean()
chicago_counts_month = pd.DataFrame(chicago_counts_month)
chicago_counts_month.columns = ['Mean']
chicago_counts_month['Min'] = chicago_counts_monthBYyear.groupby('Month')['Counts'].min()
chicago_counts_month['Max'] = chicago_counts_monthBYyear.groupby('Month')['Counts'].max()
chicago_counts_month['Std'] = chicago_counts_monthBYyear.groupby('Month')['Counts'].std()
chicago_counts_month['Mean + Std'] = chicago_counts_month['Mean'] + chicago_counts_month['Std']
chicago_counts_month['Mean - Std'] = chicago_counts_month['Mean'] - chicago_counts_month['Std']
chicago_counts_month = chicago_counts_month.reset_index()
chicago_counts_month
```

```python
#adding the empty months
for m in [1,2,6,7,12]:
    chicago_counts_month.loc[len(chicago_counts_month)] = [m,0,0,0,0,0,0]
chicago_counts_month
```

```python
#ploting counts by months over the years
sns.lineplot(data=chicago_counts_month, x='Month', y='Max', label='Max')
sns.lineplot(data=chicago_counts_month, x='Month', y='Mean + Std', label='Mean + Std')
sns.lineplot(data=chicago_counts_month, x='Month', y='Mean', label='Mean')
sns.lineplot(data=chicago_counts_month, x='Month', y='Mean - Std', label='Mean - Std')
sns.lineplot(data=chicago_counts_month, x='Month', y='Min', label='Min')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

Most collision in October, the mid autumn month, followed by May, the late spring month.
